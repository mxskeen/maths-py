# importing req libraries for vector operations

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set() #for showing interface and grid for vector coordinate systems below 
```

#plotting a vector

```python
plt.quiver(0,0,4,5) #syntax  Plot a 2D field of arrows.(x pos,y pos,x,y)
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1) #scale_units and angles are for having proper scales and units for graph
plt.xlim(-8,8)
plt.ylim(-8,8)
```

```python
#for changing color add parameter color in the above syntax
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
```

```python
#to add more than 1 vector in the same graph just repeat the first line of syntax
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,-6,0-5, scale_units='xy', angles="xy", scale=1, color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
```

for addition of vector

```python
#use numpy.asarray as we know vectors are arrays only
```

```python
vector_1= np.asarray([0,0,2,3]) #syntax for inputting vector for addition
vector_2= np.asarray([0,0,3,-2])
sum= vector_1 + vector_2
print(sum)
```

```python
#just plot the above values manually in the graph 
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy',scale=1, color='y')
plt.quiver(0,0,5,1, scale_units='xy', angles='xy',scale=1, color='r') #sum of above 2 vectors in red
plt.xlim(-8,8)
plt.ylim(-8,8)
```

for subtraction of two vectors

```python
#follow the same syntax and pattern as above just subtract in the difference
```

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
diff= vector_1 - vector_2
print(diff)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
```

multiplication of vector by a scalar

```python
vector_1=np.asarray([0,0,4,5])
vector_2=2*vector_1
print(vector_2)
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,8,10,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-12,12)
plt.ylim(-12,12)
```

```python
vector_1=np.asarray([0,0,4,5])
vector_2=-0.5*vector_1
print(vector_2)
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-2,-2.5,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-12,12)
plt.ylim(-12,12)
```
